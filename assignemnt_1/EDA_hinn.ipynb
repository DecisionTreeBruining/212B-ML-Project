```python
# count plot for all feature vairiables in the dataset
count_vis = mental_health.drop(["Timestamp"], axis = 1) # get rid of Timstamp 

fig, ((ax1, ax2), (ax3, ax4),(ax5,ax6),(ax7,ax8),(ax9,ax10)) = plt.subplots(ncols = 2,nrows = 5)
plt_1 = sns.countplot(y = count_vis['Occupation'], ax = ax1)
plt_2 = sns.countplot(y = count_vis['family_history'], ax = ax2)
plt_3 = sns.countplot(y = count_vis['treatment'], ax = ax3)
plt_4 = sns.countplot(y = count_vis['Days_Indoors'], ax = ax4)
plt_5 = sns.countplot(y = count_vis['Mental_Health_History'], ax = ax5)
plt_6 = sns.countplot(y = count_vis['Work_Interest'], ax = ax6)

# adjusting the figure size for better visual experience
fig = plt.gcf()
fig.set_size_inches(15,20)
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(ncols = 2,nrows = 2)
plt_1 = sns.countplot(y = count_vis['Growing_Stress'], ax = ax1)
plt_2 = sns.countplot(y = count_vis['Mood_Swings'], ax = ax2)
plt_3 = sns.countplot(y = count_vis['Coping_Struggles'], ax = ax3)

# adjusting the figure size for better visual experience
fig = plt.gcf()
fig.set_size_inches(10,10)
```
