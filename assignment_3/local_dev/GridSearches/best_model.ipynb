```python
import pandas as pd
from great_tables import GT, md, html, from_column, style, loc

# Data for the table
data = {
    "Hyperparameter": ["Baseline", "Number of Layers and Neurons", "Activation Function", "Initial Learning Rate", "Maximum Iterations", "Number of Iterations with No Change", "Batch Size", "Regularization Parameter (Alpha)"],
    "Value": ["Adam, 1 layer, 100 neurons, ReLU, 0.001, 200, 10, auto, 0.0001",
              "[100, 100, 100, 100, 100]", 
              "logistic", 
              "0.01", 
              "100", 
              "100", 
              "100", 
              "0.0"],
    "Recall": [0.75, 0.52, 0.66, 0.77, 0.75, 0.71, 0.74, 0.72],
    "ROC AUC": [0.81, 0.76, 0.79, 0.82, 0.81, 0.81, 0.82, 0.81],
    "Accuracy": [0.73, 0.81, 0.75, 0.73, 0.73, 0.75, 0.74, 0.75],
    "Fit Time (s)": [67, 833, 154, 73, 106, 399, 151, 102]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Create the GT table
table = GT(df)

# Apply some styling
table = table.set_style(
    style={
        loc.header(): {
            "background-color": "#f7f7f7",
            "font-weight": "bold",
        },
        loc.body(): {
            "border": "1px solid #dddddd",
            "padding": "8px",
        },
        loc.column("Hyperparameter"): {
            "text-align": "left",
            "font-weight": "bold",
        },
        loc.columns(["Value", "Recall", "ROC AUC", "Accuracy", "Fit Time (s)"]): {
            "text-align": "center",
        }
    }
)

# Display the table in Markdown format (for Jupyter Notebook or similar)
print(md(table))

# Display the table in HTML format (for web embedding)
print(html(table))
```
